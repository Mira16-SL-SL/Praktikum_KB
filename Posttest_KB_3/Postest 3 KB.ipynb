```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Breast_Cancer.csv")
```

```python
df.head()
```

### Visualisasikan dataset yang kalian miliki dan simpulkan informasi apa yang dapat dilihat dari visualisasi tersebut. Dengan ketentuan: a. Menggunakan minimal 3 tipe visualisasi data dan wajib menggunakan pie chart

## 1. Bar Plot

```python
x = list(df["Marital Status"])
y = list(df["Age"])

plt.bar(x,y,color='red')
plt.xlabel("Marital Status")
plt.ylabel("Age")
plt.title("Perbandingan Marital Status")
```

### Kesimpulan : dari bar plot diatas menampilkan data dengan title Perbandingan Marital Status. Dimana x label = marital status dan y label = age. maka output yang dihasilkan adalah marital status seperti  Married rata-rata usia (age 68), Divorced dengan rata-rata usia 68, Singgle Marital Status rata-rata usia 68, Windowed dengan rata-rata usia 68, dan terakhir ada Separated dengan rata-rata usia 67. 

## Pie Chart

```python
df["Marital Status"].value_counts()
```

```python
data_pie = df["Marital Status"].value_counts().rename_axis("Marital Status").reset_index(name="id_counts")
```

```python
data_pie.head()
```

```python
plt.figure(figsize=(10,10))
plt.pie(data_pie["id_counts"],labels=data_pie['Marital Status'],startangle=90,autopct='%.1f%%',shadow=True ,explode=(0,0,0,0,0))
plt.title('Persentase Perbandingan Marital Status')
plt.show()
```

### Kesimpulan: Dari pie chart diatas dapat kita simpulkan bahwa pie chart tersebut menampilkan persentase perbandingan marital status dalam bentuk grafik pie. Dimana marital status yang paling banyak yaitu Married dengan jumlah 2643 orang jika di persentasekan 65.7%. Cara mengetahui jumlah marital status yang belum di persentasekan di grafik pie yaitu dengan df["Marital Status"].value_counts() . Kemudian, di pie chart setelah Merried, ada Single dengan total persentase 15.3% (atau sekitar 615 orang). Selanjutnya ada Divorced dengan perbandingan 12.1% (sekitar 486 orang), lalu ada Windowed dengan persentase 12.1% ( sekitar 235 orang), dan terakhir ada separated dengan perbandingan 1.1%(sekitar 45 orang). Di pie chart ini tidak bisa menggunakan x labell dan y label.

## Scatter Plot
